```python
import numpy as np
import pandas as pd
import ipyvolume as ipv
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import cv2
import ast
from skimage.feature import hog
```

```python
dataset = pd.read_csv("Enhanced_dataset.csv")
data = dataset[['filename','region_shape_attributes','region_attributes']]
```

```python
print(data['filename'])
```

```python
type(data['region_shape_attributes'])
```

```python
#z = data.region_shape_attributes
#print(z[0])
#res = ast.literal_eval(z[0])
#print(res['x'])
x_coor = []
y_coor = []
height = []
width = []
count = 0

for z in data.region_shape_attributes:
    res = ast.literal_eval(z)
    x_coor.append(res['x'])
    y_coor.append(res['y'])
    height.append(res['height'])
    width.append(res['width'])

#print(x_coor)
Motorcycle = []
Helmet = []
for z in data.region_attributes:
    res = ast.literal_eval(z)
    Motorcycle.append(res['Motorcycle'])
    Helmet.append(res['Driver Helmet'])
    
#print(x_coor)
#print(y_coor)
#print(height)
#print(width)
#print(Motorcycle)
#print(Helmet)
```

```python
FinalDF = pd.DataFrame({'image name' : data['filename'],'x_coordinate' : x_coor,'y_coordinate' : y_coor,'height' : height,'width' : width,
                       'Motorcycle' : Motorcycle,'Helmet' : Helmet})
```

```python
hog_features = []
hog_images = []
for i in range(int(len(data['filename']))):
    path = 'Experi_imag/'+data['filename'][i]
    img = cv2.imread(path)
    #print(ssize)
    frame1 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    filter_img = cv2.GaussianBlur(frame1,(5,5),100)
    crop = filter_img[y_coor[i]:(y_coor[i]+height[i]),x_coor[i]:(x_coor[i]+width[i])]
    print(i)
    print(y_coor[i])
    print(height[i])
    print(x_coor[i])
    print(width[i])
    print(crop.shape)
    print()
    resized = cv2.resize(crop,(64,128),interpolation = cv2.INTER_AREA)
    hf,hi = hog(resized, orientations=9, pixels_per_cell=(8,8),cells_per_block=(2, 2),visualize = True)
    hog_images.append(hi)
    hog_features.append(hf)
    cv2.destroyAllWindows()


#print(hog_features)
```

```python
print(hog_features[0])
```

```python
pca = PCA(n_components = 3)
principalComponent = pca.fit_transform(hog_features)
principalDf = pd.DataFrame(data = principalComponent
             , columns = ['principal component 1', 'principal component 2', 'principal component 3'])
Motor = np.array(Motorcycle)
motor = Motor[:][:]
principalDf['Motorcycle'] = motor
Helmt = np.array(Helmet)
helmt = Helmt[:][:]
principalDf['Helmet'] = helmt
```

```python
print(principalDf)
```

```python
m = principalDf[principalDf['Motorcycle'] == 'No']
print(m)

h = principalDf[principalDf['Helmet'] == 'No'] 
print(h)
```

```python
fig = ipv.figure()
ipv.scatter(principalDf['principal component 1'],principalDf['principal component 2'],principalDf['principal component 3'])
ipv.scatter(m['principal component 1'],m['principal component 2'],m['principal component 3'],color = 'blue',size = 4)
ipv.show()
```
