```python
import pandas as pd
import numpy as np
import cv2
from skimage.feature import hog
from sklearn import metrics
```

```python
Vehicle_data = pd.read_csv('single_lane_dataset_1.csv')
```

```python
MotorCycle = Vehicle_data[Vehicle_data['MotorCycle'] == 'Yes']
```

```python
MotorCycle
```

```python
MotorCycle = np.array(MotorCycle)
```

```python
helmet = MotorCycle[:,7]
```

```python
helmet
```

# Extracting features.......

```python
hog_features = []
hog_images = []
for i in MotorCycle:
    print(i)
    path = 'Experi_imag/'+i[1]
    img = cv2.imread(path)
    vehicle = img[i[3]:(i[3]+i[4]),i[2]:(i[2]+i[5])]
    helmet_portion = vehicle[0:int(0.25 * vehicle.shape[0]) , 0:(vehicle.shape[1])]
    filter_img = cv2.GaussianBlur(helmet_portion,(5,5),100)
    resized = cv2.resize(filter_img,(64,128),interpolation = cv2.INTER_AREA)
    hf,hi = hog(resized, orientations=9, pixels_per_cell=(8,8),cells_per_block=(2, 2),visualize = True)
    hog_images.append(hi)
    hog_features.append(hf)
```

```python
hog_features = np.array(hog_features)
```

```python
var = []
for i in helmet:
    if i  == 'Yes':
        var.append(1)
    else:
        var.append(0)
var = np.array(var)
```

```python
df = pd.DataFrame(hog_features)
df['Helmet'] = var
```

```python
data = pd.read_csv('single_lane_helmet_hog_features.csv')
```

```python
data.head()
```

```python
data = np.array(data)
```

```python
features = data[:,1:-1]
variables = data[:,3781]
```

```python
var = []
for i in variables:
    if i  == 'Yes':
        var.append(1)
    else:
        var.append(0)
var = np.array(var)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features,var,test_size = 0.3, random_state = 109)
```

```python
from sklearn import svm
#creating svm classifier with kernal type as linear
clf = svm.SVC(kernel = 'poly', degree = 3)
# training the model using training set
clf.fit(x_train,y_train)
# predicting the output for test_set
y_pred = clf.predict(x_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
